```python
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('crp_prdn.csv')
```

```python
df.head()
```

1. Filling with Default Value

```python
df1 = df.fillna(value = 59)
df1.head()
```

2. Filling with Previous Value

```python
df1 = df.fillna(method = 'pad')
df1.head()
```

3. Filling with Next Value

```python
df1 = df.fillna(method = 'bfill')
df1.head()
```

4. Filling with Mean

```python
mean = int(df['Area'].mean())
df1 = df.fillna(value = mean)
df1.head()
```

5. Filling with Median

```python
df1 = df.fillna(df['Area'].median())
df1.head()
```

6. Filling with Mode

```python
df1 = df.fillna(df['Area'].mode().max())
df1.head()
```

7. Filling with Max

```python
df1 = df.fillna(df['Area'].max())
df1.head()
```

8. Filling with Min

```python
df1 = df.fillna(df['Area'].min())
df1.head()
```

9. Filling with Most Frequent

```python
df1 = df.fillna(df['Area'].mode().max())
df1.head()
```

10. Filling using KNN

```python
knn_imp = KNNImputer(n_neighbors=2)
df['Area'] = knn_imp.fit_transform(df[['Area']]).ravel()
df.head()
```

```python
#interploation
df['Area'] = df['Area'].interpolate(method='linear', limit_direction='both')
df.head()
```
