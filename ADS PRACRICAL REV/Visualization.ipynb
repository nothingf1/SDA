```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('passengerxtotal.csv')
grp_df = pd.read_csv('uber_data_group.csv')
```

```python
df.head()
```

```python
grp_df.head()
```

Bar Plot

```python
passenger = df['passenger_count']
amount = df['total_amount']

fig = plt.figure(figsize = (10, 5))

plt.bar(passenger, amount, color ='skyblue', width = 0.4)

plt.xlabel("No of Passengers")
plt.ylabel("Total Amount")
plt.title("No of Passengers and Total Cab Fare")
plt.show()
```

```python
distance = grp_df['trip_distance']
total = grp_df['total_ammount']

fig = plt.figure(figsize = (10, 5))

plt.bar(distance, total, color ='skyblue', width = 0.4)

plt.xlabel("Distance Traveled")
plt.ylabel("Total Cost")
plt.title("Distance Travel & Total Cab Fare")
plt.show()
```

Scatter Plot

```python
plt.scatter(passenger, amount)
plt.show()
```

```python
plt.scatter(distance, total)
plt.show()
```

Histogram

```python
plt.hist(total, color='skyblue', edgecolor='black')
```

Box Plot

```python
data = [passenger, amount, total]
plt.boxplot(data)
plt.show()
```

Violin Plot

```python
plt.violinplot(data)
plt.show()
```

Pie Chart

```python
plt.pie(total, labels=distance)
plt.show()
```

Heatmap

```python
hm = sns.heatmap(data = df)
plt.show()
```

Stackplot

```python
plt.stackplot(passenger, amount)
plt.show()
```
